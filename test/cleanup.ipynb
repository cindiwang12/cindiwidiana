```python
import os
from pathlib import Path

workspace = Path(r"c:\Users\Cindy\Documents\LUNARA BAKES - Copy")

# Rename directories with spaces to hyphens
old_paths = [
    (workspace / 'pages' / 'sign up', workspace / 'pages' / 'sign-up'),
    (workspace / 'public' / 'halaman utama', workspace / 'public' / 'halaman-utama'),
]

for old_path, new_path in old_paths:
    if old_path.exists():
        print(f'Renaming {old_path.name} -> {new_path.name}')
        old_path.rename(new_path)
    else:
        print(f'Skipping missing path {old_path}')

print("Folder renaming complete!")
```
